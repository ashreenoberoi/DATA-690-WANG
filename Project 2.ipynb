**Question 1: Has the employment improved over time which means have the jobs increased?**


Lets do this. Let's add all the year and employment that year and plot them to see has it increased over time

```python
dframe.groupby('Year')['Employment'].sum()
```

```python
dframe.groupby('Year')['Employment'].sum().plot(kind='bar',figsize=(50, 15), fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Employment', fontsize=20)
plt.title('Employment Over years', fontsize=20)
```

Clearly shows that the employment saw really good hike in 1990 and it kept increasing almost every year but there was a drop in 2019.


**Question 2: Which year was the unployment highest?**


Lets do this. Let's add all the year and employment that year and plot them to see has it increased over time

```python
dframe.groupby('Year')['Unemployment'].sum()
```

```python
dframe.groupby('Year')['Unemployment'].sum().plot(kind='bar',figsize=(50, 15), fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Unemployment', fontsize=20)
plt.title('Unemployment Over years', fontsize=20)
```

2010 has the most unemployment rate. 2019  even though had a drop in employment, the unemployment still remain low.Just to see if this is the same situation for the labor force, let's plot that and see

```python
dframe.groupby('Year')['Labor Force'].sum()
```

```python
dframe.groupby('Year')['Labor Force'].sum().plot(kind='bar',figsize=(50, 15), fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Labor Force', fontsize=20)
plt.title('labor Force over the years', fontsize=20)
```

The labor force saw a hike in 2010 , it means that Labor force increased and they had more people to hire than the jobs. This makes complete sense because the employment rate went down and the unemployment was the highest.

**Question 4:Are the jobs preliminary or final?**

Preliminary jobs: It means that your job is not premanent, basically on contract basis 
Final: The job is final.

Alot of the internation students are kept in the preliminary status because they don't want to sponser their H1B, I have had so many friends go through this that I actually want to see is it true? Does it happen to majority people?

```python
dframe.groupby('Status (Preliminary / Final)')['Employment'].sum()
```

Prelim and Preliminary mean the same. Let's just sum them up and put it in a pie chart to see the percentages.

```python
Total = [124802624,36604992361]
my_labels = 'Preliminary','Final'
plt.pie(Total,labels=my_labels,autopct='%1.1f%%')
plt.title('Job Status')
plt.axis('equal')
plt.show()
```
