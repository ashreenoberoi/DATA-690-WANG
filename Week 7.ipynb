```python
import os, sys
import glob
import pandas as pd
```

```python
path =os.getcwd()
print (path)
```

```python
extension = 'csv'
result = glob.glob('*.{}'.format(extension))
print(result)
```

```python
frames=[]
for i in result:
    df = pd.read_csv(i, usecols = ["UNITID", "INSTNM", "STABBR", "REGION", "ADM_RATE", "TUITIONFEE_IN"])
    print(i)
    frames.append(df)
print(frames)
```

```python
fin = pd.concat(frames)
```

```python
print (fin)
```

```python
fin1 = fin.loc[fin["INSTNM"] == 'University of Maryland-Baltimore County']
```
