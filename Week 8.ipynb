**Let's Load Libraries**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
import datetime as dt
import  time
```

**Le's load the first dataset for the president campaign**

**Question1: Which state had most president votes?**

For this we need to group by the states with total votes to sun up all the countys in every state

```python
dframe.groupby('state')['total_votes'].sum()
```

**Let's plot it and get a better picture**

```python
dframe.groupby('state')['total_votes'].sum().plot(kind='bar',figsize=(50, 15), fontsize=20)
plt.xlabel('state', fontsize=20)
plt.ylabel('total votes', fontsize=20)
plt.title('Number of votes per state', fontsize=20)
```

**This shows that california had most number of votes that were counted and second most number is for Florida**

**Which president had most votes?**

For this we need to group by the candidates and the total votes


```python
dframe.groupby('candidate')['total_votes'].sum()
```

**Let's plot it for better results**

```python
dframe.groupby('candidate')['total_votes'].sum().plot(kind='bar',figsize=(50, 15), fontsize=20)
plt.xlabel('candidates', fontsize=20)
plt.ylabel('total votes', fontsize=20)
plt.title('Number of votes per candidate', fontsize=20)
```

**Joe Biden has more votes**
